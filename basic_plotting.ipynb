Here we are going to check another way to plot using a different backend 

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure

fig = Figure() # and associate it with the backend
canvas = FigureCanvasAgg(fig)
ax = fig.add_subplot(111)
ax.plot(3, 2, ".")
# but jupyter can't render it directly, so
canvas.print_png("test.png")
```

```python
# let's create another chart with pyplot
plt.figure()
plt.plot(3, 2, "o")
ax = plt.gca()
ax.axis([0,6,0,10]) # here we are limitating the min and max values of the axes
```

```python
# here it will add more data to our chart, the points are render with different colors
# as the axes recognices them as different data series
plt.figure()
plt.plot(1.5, 1.5, "o")
plt.plot(2, 2, "o")
plt.plot(2.5, 2.5, "o")
```
