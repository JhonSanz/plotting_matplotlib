```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas_data = pd.read_csv("gas_prices.csv")
gas_data.head()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Gas price over time (in USD)")
plt.plot(gas_data["Year"], gas_data["USA"], "r.-", label="USA")
plt.plot(gas_data["Year"], gas_data["Canada"], "g.-", label="Canada")
plt.plot(gas_data["Year"], gas_data["South Korea"], "b.-", label="South Korea")

plt.legend()

# print(gas_data["Year"][::2]) # We used it to avoid overlapped labels
# it means that the xticks are taken every 2 years
plt.xticks(gas_data["Year"][::2])
plt.xlabel("Year")
plt.ylabel("USD")
plt.show()
```

```python
# Obviously you can plot it with a loop
plt.figure(figsize=(8, 5))
plt.title("Gas price over time (in USD)")

for item in gas_data.columns[1:]:
    plt.plot(gas_data["Year"], gas_data[item], ".-", label=item)

plt.legend()
plt.xticks(gas_data["Year"][::2])
plt.xlabel("Year")
plt.ylabel("USD")
plt.show()
```

```python
# Now let's move the legend outside the chart, at the right side.
plt.figure(figsize=(10, 6))
ax = plt.subplot(1, 1, 1)
ax.set_title("Gas price over time (in USD)")

for item in gas_data.columns[1:]:
    ax.plot(gas_data["Year"], gas_data[item], ".-", label=item)

ax.legend(loc="center left", bbox_to_anchor=(1, 0.5))
ax.set_xticks(gas_data["Year"][::2])
ax.set_xlabel("Year")
ax.set_ylabel("USD")

# bbox_inches="tight" avoids cutting the chart after saving it
plt.savefig("gas_price_figure.png", bbox_inches="tight", dpi=300)
plt.show()
```
