```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# We can add the dimensions of the figure to get a better image quality
# dpi: The resolution of the figure in dots-per-inch.
# plt.figure(figsize=(5, 3), dpi=300)
plt.figure()

# to enable .legend() we have to add the label property
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(
    x, y, label="2x",
    color="#0aFb55", linewidth=5,
    marker=".", markersize=20, markerfacecolor="black",
    markeredgecolor="red", linestyle=":"
)

# BUT there is a shorthand notation
# ftm = '[color][marker][line]'
x = np.arange(0, 4.5, 0.5)
plt.plot(x[:6], x[:6] ** 2, 'r.-', label="x$^2$")
plt.plot(x[5:], x[5:] ** 2, 'r.:')


# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html
plt.title(
    "First graph :D",
    fontdict={
        "fontname": "Comic Sans MS",
        "fontsize": 20,
        "color": "red"
    }
)
plt.xlabel("My x axis")
plt.ylabel("My y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7.5,8,10]) # it keeps that space (7, 7.5), it could be useful

plt.legend()

plt.savefig("myfirstgraph.png", dpi=300)

plt.show()
```
