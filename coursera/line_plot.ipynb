```python
import numpy as np
import matplotlib.pyplot as plt

linear_data = np.arange(1,9)
quadratic_data = linear_data ** 2
cubic_data = linear_data ** 3
plt.figure()
plt.plot(linear_data, "-^", quadratic_data, "-o",  cubic_data, "-o")
# Notice that here we dont need to specify the x values,
# and here it understands that both are different classes
# and it paints it different.
# you can add a series like this
plt.plot([22, 44, 55], "--r")
plt.xlabel("some data")
plt.ylabel("some other data")
plt.title("A little")
plt.legend(["Baseline", "Competition", "Us", "Something"])
```

```python
# now we want to add a shadow between to clases,
# it could be useful to represent some characteristics
# on the data
plt.gca().fill_between( # gca means "get current axes"
    range(len(linear_data)), linear_data, quadratic_data,
    facecolor="blue", alpha=0.2
)
```

Now let's change our x axis to date intervals

It is important to know that the datetime types are usually difficult to deal with it, so here we are going to use the pandas library to convert correctly the datetime strings to datetime objects

```python
import pandas as pd

plt.figure()
observation_dates = np.arange("2017-01-01", "2017-01-09", dtype="datetime64[D]")
observation_dates = list(map(pd.to_datetime, observation_dates))
plt.plot(
    observation_dates, linear_data, "-o",
    observation_dates, quadratic_data, "-o"
)
# the dates are overlapping so
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)

# here, look that we are accesing to the axes directly and indirectly
ax = plt.gca()
ax.set_xlabel("Date")
plt.ylabel("Units")
# you can use latex format like this
plt.title("Quadratic ($x^2$) vs Linear ($x$) performance")
```
