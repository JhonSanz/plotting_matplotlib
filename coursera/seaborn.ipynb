```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name="v1", dtype=float)
v2 = pd.Series(2 * v1 + np.random.normal(60, 15, 1000), name="v2", dtype=float)
```

```python
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50, 150, 5), label="v1")
plt.hist(v2, alpha=0.7, bins=np.arange(-50, 150, 5), label="v2")
plt.legend()
```

```python
plt.figure()
plt.hist([v1, v2], histtype="barstacked", density=True)
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3, hist_kws={"color": "Teal"}, kde_kws={"color": "Navy"})
```

```python
sns.jointplot(data={"v1": v1, "v2": v2}, x="v1", y="v2", alpha=0.4, kind="hex")
```

```python
sns.set_style("white")
sns.jointplot(data={"v1": v1, "v2": v2}, x="v1", y="v2", space=0, kind="kde")
```
