```python
import numpy as np
import matplotlib.pyplot as plt

linear_data = np.arange(1,9)
quadratic_data = linear_data ** 2
plt.figure()
xvals = range(len(linear_data))
print(xvals)
plt.bar(xvals, linear_data, width=0.3)
# and if you add a new class it will add it to the same axis value
# but with different color
plt.bar([item + 0.3 for item in xvals], quadratic_data, width=0.3)
```

```python
# finally something useful is to stack the bars
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color="r")
plt.bar(xvals, quadratic_data, width=0.3, bottom=linear_data, color="g")
```
