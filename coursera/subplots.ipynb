```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
# 1 row, 2 columns, 1 plotnumber
plt.subplot(1, 2, 1)
linear_data = np.arange(1, 9)
plt.plot(linear_data, "-o")

# Here number two means that we are going to add 
# the next plot in the second position of the figure
plt.subplot(1, 2, 2)
exponential_data = linear_data ** 2
plt.plot(exponential_data, "-o")
```

Notice that both have different dimentions, and the scale is not the same for each.

It would be great to keep the same y-scale axis for both

```python
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, "-o")

ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.plot(exponential_data, "-x")
```

```python
# Going further you can get great results.

# NOTICE THAT exist both, .subplot and .subplots
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True,  sharey=True)
ax5.plot(linear_data, "-")
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True,  sharey=True)
ax5.plot(linear_data, "-")

# gcf means "get current figure"
# it's supposed that doing this you can restore the labels in each subplot
for ax in plt.gcf().get_axes():
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible(True)

plt.gcf().canvas.draw()
```
