```python
import matplotlib.pyplot as plt
import numpy as np
```

Using heatmaps you can get great information from the data, but it has to be used in specific use cases.

```python
plt.figure()

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)

_ = plt.hist2d(X, Y, bins=25)
```

```python
plt.figure()

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
# By increasing the bins the result looks much more difuse
_ = plt.hist2d(X, Y, bins=100)

# add the beuatiful legend
plt.colorbar()
```
